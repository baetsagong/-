```python
import pandas as pd
# df = pd.read_csv("members.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch2/members.csv")
```

```python
std = df['age'].std()
mean = df['age'].mean()
print(std)
print(mean)
```

```python
lower = mean - (std * 1.5)
upper = mean + (std * 1.5)
print(lower)
print(upper)
```

```python
cond1 = df['age'] < lower
cond2 = df['age'] > upper
print(cond1)
print(cond2)
```

```python
print(df[cond1|cond2]['age'].sum())
```
